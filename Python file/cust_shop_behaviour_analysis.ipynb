> 1] Now as we can see there are 37 missing values in "Review Rating" column, basic logic we will use here is to fill by avg median of column by adjoining category and Item Purchased columns.\
> 2] Also we will group these both column together to calculate mean and fill values in "Review Rating"column.

```python
df["Review Rating"]= df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

> In this way all missing values are filled up by selecting category with median to transform.

> Now the next part is if we look at part of columns name, those are not constant some has spaces, capitalcase, smallcase, etc.

> It is always good have practice to use snakecase; we can refer code in SQL and python, it is also readable

```python
df.columns= df.columns.str.replace(" ","_")
df.columns= df.columns.str.lower()
df = df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

> Now lets create age category i.e by refering to age column 

```python
labels= ["Young adult","Adult","Middle-Age","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4, labels=labels)
#qcut:-It splits numerical data into bins so that each bin contains 
#approximately the same number of observations.
#It splits numerical data into bins so that each bin contains
```

```python
df[["age","age_group"]].head(8) 
```

> If we look at column"frequency_of_purchases" it is in text so again we can add column and create numerical frequency column. 

```python
frequency_count= {         #Create dictionary which maps each textual frequcny = no.of days
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annualy":365,
    "Every 3 Months":90
}

df["purchased_frequency_days"]=df["frequency_of_purchases"].map(frequency_count)
```

```python
df[["purchased_frequency_days","frequency_of_purchases"]].head()
```

> Now if we look at columns "discount_applied","promo_code_used" both have similar textual data.\
But let's check if both columns are actually same.

```python
(df['discount_applied']==df['promo_code_used']).all()
```

> So now we know both columns are exactly same, let's drop any one column 

```python
df= df.drop("promo_code_used",axis=1)
```
